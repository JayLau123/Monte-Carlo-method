```python
import random
import matplotlib.pyplot as plt
import math
import statistics
```

The length of one needle: a

The width of parallel lines: d

Since the needle is randomly thrown on the table, a two-dimensional random variable (X,Y) is used to determine its specific position on the table. X represents the distance between the midpoint of the needle and the parallel line, and Y represents the angle between the needle and the parallel line.

X has a uniform distribution in the range of $(0, \frac{d}{2})$, Y has a uniform distribution in the range of $(0, \frac{\pi}{2})$. X and Y is independent of each other. The probability density function of (X,Y) can be written as follows:

if $x<\frac{a}{2}sin(Y)$, the needle intersects the nearest parallel line. 

$$f(x,y)=
\begin{cases}
\dfrac{4}{\pi d} &, {0<x<\frac{d}{2}, 0<y< \frac{\pi}{2}} \\
0 &, {other}
\end{cases}$$

$$P(X<\frac{a}{x}sin(Y))=\iint f(x,y)dxdy=\int_{0}^{\dfrac{\pi}{2}}\int_{0}^{\dfrac{a}{2}sin(Y)}\dfrac{4}{\pi d}dxdy=\frac{2a}{\pi d}$$



How many needles to throw: N

How many needles intersects the parallel lines: M 

The probability is $\dfrac{M}{N}$


Therfore, $\dfrac{M}{N}=\dfrac{2a}{\pi d},    \pi=\dfrac{2aN}{dM}$


```python
def Monte_Carlo(runs,needles,a,d):
    pi_values = [] # store different value of PI

    plt.axhline(y=math.pi, color='r', linestyle='-') #Horizontal line for actual value of PI :
    
    for i in range(runs):
        
        nhits = 0 
        
        for j in range(needles): # Initialize the position of all needles
          
            x = random.uniform(0,d/2.0)   # min = 0  max = b_width/2, the distance between the midle point of needle and the nearest line.
            theta = random.uniform(0,math.pi/2)   # the angle will be from 0 to pi/2 
            
            
           
            xtip = x - (a/2.0) * math.cos(theta)  # check if the needle crosses the line or not :
            if xtip < 0 :
                nhits +=1
    
    
        
        numerator = 2.0 * a * needles  # formula : pi=numerator/denominator
        denominator = d * nhits
        pi_values.append((numerator/denominator))   #Append the value of pi 
    
    #print(pi_values[-1]) # find the last pi value after all iterations
    
    average_pi=statistics.mean(pi_values) #  the average value of pi
    print("Average value of pi",average_pi)
    
   # print(sum(pi_values))
    plt.plot(pi_values)
    
# total number of needles and runs 
runs = 200
needles = 1000

a = 2 #Length of needle
d = 2 # distance between 2 parallel lines>=length of needle

Monte_Carlo(runs,needles,a,d)
```
